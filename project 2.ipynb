```python
import pandas as pd
filepath =  'C:\\Users\\mgane\\Desktop\\problem solving/AppleStore.csv'
def readdata(filepath):
    return pd.read_csv(filepath)
fp = readdata(filepath)
pf = fp[fp['prime_genre']== 'Games']
ps = fp[fp['prime_genre']!= 'Games']
avggame =pf['user_rating'].sum()//len(pf['user_rating'])
avgng = ps['user_rating'].sum()//len(ps['user_rating'])
print ('The avg of ratings of gaming apps is ',avggame)
print ('The avg of ratings of non-gaming apps is ',avgng)
```

```python
import matplotlib.pyplot as pt
ctype=['Free', 'Affordable','Expensive','Very Expensive']
(free,aff,exp,vexp)=(0,0,0,0)
for i in fp['price']:
    if i<50:
        if i<20:
            if i==0:
                free+=1
            aff+=1
        exp+=1
    vexp+=1
counts = [free,aff,exp,vexp]
pt.pie(counts,labels= ctype,shadow=True,startangle=90,autopct='%1.1f%%')
pt.legend(loc='best')
pt.show()
print (counts)


```
